```python
import pandas as pd
```

```python
df = pd.read_csv('cases/weather.csv')
df['Date'] = pd.to_datetime(df['Date'])
# make date values to unix timestamp
df['Date'] = df['Date'].astype('int64') // 10**9
df
```

```python
timeseries = df['Date']
timeseries
```

```python
airfields = df.T.iloc[1:, :]
airfields
```

```python
minor = airfields.iloc[0, :].T
minor
```

```python
minor_df = pd.DataFrame({'Airfield': 'AD', 'Date': timeseries, 'Weather': minor})
minor_df
```

```python
major_df = pd.DataFrame(columns=['Airfield', 'Date', 'Weather'])
for i in range(0, len(airfields)):
    major_df = major_df._append(pd.DataFrame({'Airfield': airfields.index[i], 'Date': timeseries, 'Weather': airfields.iloc[i, :].T}))
```

```python
major_df
```

```python
major_df.to_csv('cases/weather_major.csv', index=False)
```
