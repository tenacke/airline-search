```python
import os
for file in os.listdir('cases/data'):
    if file.endswith('.csv'):
        data = pd.read_csv('cases/data/' + file)
        # data = data.drop(columns=['continent', 'iso_region'])
        data.columns = ['AirportCode', 'AirfieldName', 'Latitude', 'Longitude', 'ParkingCost']
        parkCost = np.random.random((data.shape[0], 1))
        parkCost = parkCost * 200 + 800
        parkCost = parkCost.astype(dtype=np.float64)
        data['ParkingCost'] = parkCost
        data.to_csv('cases/data/' + file, index=False)
```
